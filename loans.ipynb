# Simple data plot of the data
- This clearly shows that the loan amount is decreasing over time <br>
- Taking in assumption that this is data for different customers repayment plan due to the numerous similar dates in the data, we would like to know the trends in different years and trends in the same month across different years

# loans dates in 2019

```python

df_2019=data[data['date'].dt.year==2019]

df_2019=df_2019.drop(columns=['date'])

df_2019=df_2019.head(10)
#df_2019=df_2019.rename(columns = {'loan_amount':'2019'},  inplace = True)
df_2019.columns = ['2019']
yr_2019 = df_2019.set_index('2019')
yr_2019.reset_index(inplace=True)
yr_2019
```

# loans dates in 2020

```python
df_2020=data[data['date'].dt.year==2020]
df_2020=df_2020.drop(columns=['date'])
df_2020=df_2020.head(9)
df_2020.columns = ['2020']
yr_2020 = df_2020.set_index('2020')
yr_2020.reset_index(inplace=True)
yr_2020
```

# loans dates in 2021

```python
df_2021=data[data['date'].dt.year==2021]
df_2021=df_2021.drop(columns=['date'])
df_2021=df_2021.head(9)
df_2021.columns = ['2021']
yr_2021 = df_2021.set_index('2021')
yr_2021.reset_index(inplace=True)

yr_2021
```

```python
years = pd.concat([yr_2019, yr_2020, yr_2021], axis=1)
```

```python
years
```

# Graph 

```python
# get columns to plot
columns = years.columns
# create x data
x_data = range(0, years.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, years[column], label=column)
# set title and legend
ax.set_title('Cost of loans over Years')
ax.legend()
```

# Comparing the loan amount over years
Loan amount decreased hitting the lowest amount in 2021 <br>




```python
df_aug=data[data['date'].dt.month==8]
df_aug=df_aug.drop(columns=['date'])
#df_2019=df_2019.head(10)
df_aug.columns = ['August']
df_aug = df_aug.set_index('August')
df_aug.reset_index(inplace=True)
df_aug
```

```python
df_sept=data[data['date'].dt.month==9]
df_sept=df_sept.drop(columns=['date'])
df_sept.columns = ['September']
df_sept = df_sept.set_index('September')
df_sept.reset_index(inplace=True)
df_sept
```

```python
df_jan=data[data['date'].dt.month==1]
df_jan=df_jan.drop(columns=['date'])
#df_2019=df_2019.head(10)
df_jan.columns = ['January']
df_jan = df_jan.set_index('January')
df_jan.reset_index(inplace=True)
df_jan
```

```python
common_months = pd.concat([df_aug, df_sept, df_jan], axis=1)
common_months
```

# Comparing loan cost of similar months across the three years, we want to know if similar factors apply will the loan cost be same

```python
# get columns to plot
columns = common_months.columns
# create x data
x_data = range(0, common_months.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, common_months[column], label=column)
# set title and legend
ax.set_title('Loan Cost of Same Month across different years')
ax.legend()
```
